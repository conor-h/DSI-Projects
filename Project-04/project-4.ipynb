```python
# Boxplot of the median ages for each class
plt.subplots(figsize=(12,6));
sns.boxplot(x = 'Pclass', y = 'Age', data = df);
plt.title('Age distribution by class\n');
plt.ylim(0,90);
```

We can see that the median age for 1st class is higher than the other classes. We can calculate the median ages for each class.

```python
# Display the median age for each class using groupby
median_ages = df.groupby('Pclass')['Age'].median().to_frame().reset_index()
median_ages
```

I will fill in the missing age values with the median of it's class.

```python
# Define a function to take the index of a row in the dataframe and return the median age for it's class
def get_median_age(indexer):
    try:
        return median_ages['Age'][(median_ages['Pclass'] == df.ix[indexer,'Pclass'])]
    except:
        return 'Failed'
```

```python
# Determine indices of the rows with missing age values
missing_ages = df[df.Age.isnull()].index.tolist()
```

```python
# Loop through the missing age values and set to the median age for it's class 
for i in missing_ages:
    df.set_value(index = i, col = 'Age', value = get_median_age(i))
```

#### 10. What do the classification metrics tell us?

**Precision**

The precision is a ratio of true positives against the sum of true positives and false positives. The precision score reaches its best value at 1 and worst at 0.
- _TP/(TP+FP)_

It tells us how many of the passengers predicted to survive actually survived. Our model had a precision score of 0.80. This is a good score for precision.


**Recall**

The recall is a ratio of true positives against the sum of true positives and false negatives. The recall score reaches its best value at 1 and worst at 0.
- _TP/(TP+FN)_

It tells us of all the passengers that survived, how many were predicted correctly. Our model had a recall score of 0.80. This is a good score for recall.


**F1 -score**
The F1-score is a weighted average of the precision and recall scores. The F1-score reaches its best value at 1 and worstat 0.
- _2 x (Precision x Recall) / (Precision + Recall)_

The F1-score for our model is 0.80. This is a good F1-score.

<br />

Reviewing the classification report we can see that the model peforms well, with a overall score 0.80. Looking at the scores for passengers that survived and those that didn't survice, it is clear that the model is better at predicting when a passenger didn't survive.

#### 11. Check the Confusion Matrix

```python
# Define a function for plotting confusion matrix
def plot_confusion_matrix(cm, classes, title='Confusion matrix for Titanic survival\n', cmap = plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.ylabel('Actual survival\n')
    plt.xlabel('\nPredicted survival')
    plt.show()
    return
```

```python
# Plot a confusion matrix for the test set
cnf_matrix = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cnf_matrix, classes = ['Didn\'t survive', 'Survived'])
```

#### 12. What does the Confusion Matrix tell us? 

The confusion matrix is a method of describing the performance of a classification model on a test data set when the true values are known. Predicted values are columns and the true/actual values are rows.

#### 3. Explain the difference between the difference between the L1 (Lasso) and L2 (Ridge) penalties on the model coefficients.

Both L1 and L2 regression add an additional component or weight to the loss function, to reduce the risk of overfitting.

**L1 (Lasso)**


- Adds the sum of the _absolute value_ of the $\beta$ coefficients to the residual sum of squares (RSS)

$$\lambda_1\sum_{j=1}^p |\beta_j|$$


L1 regression penalties will shrink coeffecients and zeros out the coefficients of collinear variables. Along with shrinking coefficients, L1 also performs feature selection.

<br />

**L2 (Ridge)**


- Adds the sum of the squared (non-intercept!) $\beta$ values to the residual sum of squares (RSS)

$$\lambda_2\sum_{j=1}^p \beta_j^2$$


L2 regression does not zero out coeffecients but reduces the magnitude of the coeffecients gradually and reducing model complexity.

#### 4. What hypothetical situations are the Ridge and Lasso penalties useful?

- L1 (Lasso) penalties are useful when there are high numbers of features as coeffectients reduce to zero and can be dropped from the model

- L2 (Ridge) penalties are useful to prevent overfitting in a model. As it does not drop coeffecients it is not good for cases with high numbers of features

Both L2 and L2 penalties are useful for handling outliers in a regression.

#### 5. Explain how the regularization strength (C) modifies the regression loss function. Why do the Ridge and Lasso penalties have their respective effects on the coefficients?

The regularization strength (C), is a function of the inverse alpha from the Ridge and Lasso regularisation functions.

$$C = 1/\lambda$$

From this we can see that the larger C becomes the the weaker the penalty effect on regularisation. Conversely as C goes to 0 the higher the penalty effect on the regularisation.

#### 3. How does the number of neighbors affect the bias-variance tradeoff of your model?


As we increase the number of neighbours, the variance in our model decreases and the bias increases.

Having a low number of neighbours would give higher weighting to outliers. As you increase the number of nearest neighbours you increase the chance that there will more of the majority class included.

At a sufficiently high number of nearest neighbours, the result identified will be in the majority class.
