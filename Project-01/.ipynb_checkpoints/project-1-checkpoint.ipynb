# Project 1 - Analysis of SAT Scores

- The data describes the mean SAT scores for students in each state in the United States in the year 2001. The SAT is a standardized test widely used for college admissions in the United States.

- The data includes rate of participation, mean SAT Verbal and Math scores.

|Column Name        |Description                            |
| -------------- |-------------------------------------|
|State           |US State that the SAT was taken            |
|Rate            |Rate of participation (percentage) in taking the SAT         |

### Exploratory data analysis

```python
# Set the plot size
plt.figure(figsize=(8,5))

# Plot the histogram of rate of participation
sns.distplot(df['Rate'], color="green", bins=10)
plt.ylabel('Frequency\n')
plt.xlabel('\nRate (%)')
plt.title('Distribution of the Rate\n', size = 16)
plt.grid(False)
plt.show()
```

```python
# Set the plot size
plt.figure(figsize=(8,5))

# Plot the histogram of math score
sns.distplot(df['Math'], color="blue", bins=10)
plt.ylabel('Frequency\n')
plt.xlabel('\nMath Score')
plt.title('Distribution of the Math Score\n', size = 16)
plt.grid(False)
plt.show()
```

```python
# Set the plot size
plt.figure(figsize=(8,5))

# Plot the histogram of verbal score
sns.distplot(df['Verbal'], color="orangered", bins=10)
plt.ylabel('Frequency\n')
plt.xlabel('\nVerbal Score')
plt.title('Distribution of the Verbal Score\n', size = 16)
plt.grid(False)
plt.show()
```

Looking at the histograms it can be seen that none of the histograms are normally distributed but some are closer to normal to. For example, the distribution of math scores looks the closest to being normally distributed.


We can test this using the *scipy.stats.mstats.normaltest* function which tests whether a sample differs from a normal distribution.

The function tests the null hypothesis that a sample comes from a normal distribution. It returns two values, the chi-squared statistic, and the associated p-value. The null hypothesis here is that the data came from a normal distribution, the p-value represents the probability that a chi-squared statistic that large would be seen. A very small p-valus means it is unlikely that the data came from a normal distribution.

```python
print 'Rate \n Chi-squared statistic:', stats.normaltest(df['Rate'])[0],  'p-value:', stats.normaltest(df['Rate'])[1], '\n'
print 'Math \n Chi-squared statistic:', stats.normaltest(df['Math'])[0],  'p-value:', stats.normaltest(df['Math'])[1], '\n'
print 'Verbal \n Chi-squared statistic:', stats.normaltest(df['Verbal'])[0],  'p-value:', stats.normaltest(df['Verbal'])[1]
```

After using *scipy.stats.mstats.normaltest* we can see that:

- for **Rate** the p-value is very low (1.72020028249e-18), we can confidently reject the null hypothesis and so it is not normally distributed
- for **Math** the p-value is high (0.766), we cannot reject the null hypothesis in this instance
- for **Verbal** the p-value is low (1.78030555519e-05), we can can reject the null hypothesis again and so it is also not normally distributed

```python
# Set the plot size
plt.figure(figsize=(8,5))

# Plot a scatterplot of math score vs rate
plt.scatter(df['Rate'], df['Math'], s=75, color='blue', alpha=0.5)
plt.title('Mean math score vs rate\n', fontsize=16)
plt.ylabel('Math score\n')
plt.xlabel('\nRate')
plt.grid(False)
plt.show()
```

```python
# Set the plot size
plt.figure(figsize=(8,5))

# Plot a scatterplot of math score vs rate
plt.scatter(df['Rate'], df['Verbal'], s=75, color='red', alpha=0.5)
plt.title('Mean verbal score vs rate\n', fontsize=16)
plt.ylabel('Verbal score\n')
plt.xlabel('\nRate')
plt.grid(False)
plt.show()
```

```python
# Set the plot size
plt.figure(figsize=(8,5))

# Plot a scatterplot of math score vs rate
plt.scatter(df['Rate'], df['Verbal'], s=75, color='red', alpha=0.5)
plt.scatter(df['Rate'], df['Math'], s=75, color='blue', alpha=0.5)
plt.title('Mean verbal/math scores vs rate\n', fontsize=16)
plt.ylabel('Verbal/math scores\n')
plt.xlabel('\nRate')
plt.grid(False)
plt.show()
```

There is a negative correlation between both math and verbal scores to rate, the higher the rate of participation of a state the lower the mean scores get for both math and verbal.

There also appears to be a strong positive correalation between math and verbal scores (with the exception of one outlier). This means that students who perform well in math, on average, also perform well in verbal but this is to be expected.

The correlations are:

```python
print('Correlation between Rate and Math:', np.corrcoef(df['Rate'], df['Math'])[0][1])
print('Correlation between Rate and Verbal:', np.corrcoef(df['Rate'], df['Verbal'])[0][1])
print('Correlation between Math and Verbal:', np.corrcoef(df['Math'], df['Verbal'])[0][1])
```
